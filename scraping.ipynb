```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Opening html
with open('test.html', 'r') as f:
    html = f.read()
```

```python
# Parsing html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Printing parsed soup object
print(soup.prettify())
```

```python
# getting title
soup.find('title')
```

```python
soup.find('title').text
```
